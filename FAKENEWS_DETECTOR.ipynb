```python
import pandas as pd
import numpy as np
import re
```

```python
true = pd.read_csv('/content/sample_data/True.csv',index_col=0)
true.head()
```

```python
true['label']=1
fake['label']=0
```

```python
fake =pd.read_csv('/content/sample_data/Fake.csv')
fake.head()
```

```python
news=pd.concat([true,fake],axis=0)
news.head()
```

```python
news.tail()
```

```python
news=news.sample(frac=1)
```

```python
news.drop(['title','subject','date'],axis=1)
```

```python
news.reset_index(inplace=True)
news.head()
```

```python
def wordopt(text):
  text=text.lower()
  text=re.sub(r'http>://\S+|www\.\S+','',text)
  text = re.sub(r'[^\w\s]', '', text)
  text = re.sub(r'\d+', '', text)
  text = re.sub(r'<.*?>', '', text)
  text = re.sub(r'\n', ' ', text)
  return text

```

```python
news['text']=news['text'].apply(wordopt)
```

```python
x=news['text']
y=news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization =TfidfVectorizer()
```

```python
xv_train=vectorization.fit_transform(x_train)
```

```python
xv_test=vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR =LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
LR.predict(xv_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
DTC =DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc=DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dtc))
```

```python
def output_label(n):
  if n==0:
    return "it is a fake news"
  elif n==1:
    return "news is true"
```

```python
def manual_testing(news):
  testing_news={'test':[news]} #defining a dictionary
  new_def_test=pd.DataFrame(testing_news)
  new_def_test['test']=new_def_test['test'].apply(wordopt)
  new_x_test=new_def_test['test']
  new_xv_test=vectorization.transform(new_x_test)
  pred_lr=LR.predict(new_xv_test)
  return output_label(pred_lr[0])
```
